## Hypothesis Testing on Registered Rides

you are interested                  
in identifying whether there is a statistically significant difference between                
registered rides during working days and the weekend     
In other words : - 

#### H_0 : average registered rides over weekdays - average registered rides over weekend=0

#### H_a : H_a : average registered rides over weekdays - average registered rides over weekend≠0

```python
# Once the null hypothesis is established, collect data for the two groups
# and perform the statistical test
weekend_days = ['saturday', 'sunday']
weekday_mask = prprocessed_df['weekday'].isin(weekend_days)
work_days_mask = ~prprocessed_df.weekday.isin(weekend_days)
```

```python
# select registered rides for the weekend and working days
weekend_rides = prprocessed_df[weekday_mask].registered
work_days_rides = prprocessed_df[work_days_mask].registered
```

```python
# Perform the two-sample t-tests by using the scipy.stats ttest_ind function
# The ttest_ind function performs a paired t-test for the null hypothesis that
# the two samples are drawn from the same distribution.
from scipy.stats import ttest_ind
test_result = ttest_ind(weekend_rides, work_days_rides)
print(f"Statistic value: {test_result[0]:.03f}, p-value: {test_result[1]:.03f}")
```

The resulting p-value from this test is less than 0.0001, which is far below the            
standard critical 0.05 value. As a conclusion, we can reject the null hypothesis            
and confirm that our initial observation is correct: that is, there is a statistically           
significant difference between the number of rides performed during working             
days and the weekend

```python
# now let's see the distribution of the two sample 
plt.figure(figsize=(20,10))
sns.distplot(weekend_rides)
sns.distplot(work_days_rides)
plt.legend(['weekend', 'work days'])
plt.xlabel('rides')
plt.title('Rides distributions')
```

Perform the same type of hypothesis testing to validate the second assumption            
from the last section— that is, casual users perform more rides during the           
weekend. In this case, the null hypothesis is that the average number of rides       
during working days is the same as the average number of rides during the         
weekend, both performed only by casual customers.            

```python
# now let's make the same thing for the casual users
weekend_rides = prprocessed_df[weekday_mask].casual
work_days_rides = prprocessed_df[work_days_mask].casual
test_result = ttest_ind(weekend_rides, work_days_rides)
print(f"Statistic value: {test_result[0]:.03f}, p-value: {test_result[1]:.03f}")
```

he p-value returned from the previous code snippet is 0, which is strong          
evidence against the null hypothesis. Hence, we can conclude that casual        
customers also behave differently over the weekend              

```python
# now let's see the distribution of the two sample
plt.figure(figsize=(20,10))
sns.distplot(weekend_rides)
sns.distplot(work_days_rides)
plt.legend(['weekend', 'work days'])
plt.xlabel('rides')
plt.title('Rides distributions')
```

In conclusion, we can say that there is a statistically significant difference between          
the number of rides on working days and weekend days for both casual and registered customers

```python
# i want to know if the casual user performed more rides in the weekend or in the work days
we = prprocessed_df[weekday_mask].casual.sum()
wd = prprocessed_df[work_days_mask].casual.sum()
print(f"weekend: {we}, work days: {wd}")
```

## Analysis of Weather-Related Features

as both our variables of interest (the casual and registered number of rides)           
are also continuously distributed, the first and most common way to measure the relationship             
between two different continuous variables is to measure their correlation        

```python
# let's create function to calulate the correlation between two variables
def correlation(data, col):
    corr = np.corrcoef(data[col], data["registered"])[0, 1]
    # now let's make a regplot to see the correlation
    ax = sns.regplot(x=col, y="registered", data=data,scatter_kws={"alpha":0.05},label=f"Registered rides correlation : {corr:.2f}")

    corr = np.corrcoef(data[col], data["casual"])[0, 1]
    # now let's make a regplot to see the correlation
    ax = sns.regplot(x=col, y="casual", data=data,scatter_kws={"alpha":0.05},label=f"Casual rides correlation : {corr:.2f}")
    legend = ax.legend()
    for lh in legend.legendHandles:
        lh.set_alpha(0.5)
        
    ax.set_ylabel('rides')
    ax.set_title(f"Correlation between {col} and registered/casual rides")
    return ax




```

```python
# Applying the previously defined function to the four columns (temp, atemp, hum,
# and windspeed) returns the following figure
plt.figure(figsize=(20,10))

plt.subplot(2, 2, 1)
correlation(prprocessed_df, 'temp')

plt.subplot(2, 2, 2)
correlation(prprocessed_df, 'atemp')

plt.subplot(2, 2, 3)
correlation(prprocessed_df, 'hum')

plt.subplot(2, 2, 4)
correlation(prprocessed_df, 'windspeed')
```

we can observe that higher temperatures have a positive impact           
on the number of rides (the correlation between registered/casual rides and temp is              
0.335 and 0.46, respectively, and it's a similar case for atemp)             

The same pattern can be observed in Figure 1.21, in which the humidity level has a negative

This means that with a high level of humidity (mist or rain), customers will tend not

we can see that there is minimal

## Time Series Analysis

```python
# define function for plotting rolling statistics and ADF test for time series
from statsmodels.tsa.stattools import adfuller
def test_stationarity(ts, window=10, **kwargs):
    # create dataframe for plotting
    plot_data = pd.DataFrame(ts)
    plot_data['rolling_mean'] = ts.rolling(window).mean()
    plot_data['rolling_std'] = ts.rolling(window).std()
    # compute p-value of Dickey-Fuller test
    p_val = adfuller(ts)[1]
    ax = plot_data.plot(**kwargs)
    ax.set_title(f"Dickey-Fuller p-value: {p_val:.3f}")
```

```python
daily_rides = prprocessed_df[["dteday", "registered", "casual"]]
daily_rides = daily_rides.groupby("dteday").sum()
daily_rides
```

```python
# convert index to DateTime object
daily_rides.index = pd.to_datetime(daily_rides.index)
daily_rides
```

```python
# apply the test_stationarity to the daily_rides registered
test_stationarity(daily_rides["registered"], window=10, figsize=(20,10))
```

```python
# apply the test_stationarity to the daily_rides casual
test_stationarity(daily_rides["casual"], window=10, figsize=(20,10))
```

From the performed tests, we can see that neither the moving average nor standard           
deviations are stationary. Furthermore, the Dickey-Fuller test returns values of 0.355            
and 0.372 for the registered and casual columns, respectively. This is strong         
evidence that the time series is not stationary, and we need to process them in order            
to obtain a stationary one             

```python
# A common way to detrend a time series and make it stationary is to subtract either its
# rolling mean or its last value
# subtract rolling mean
registered = daily_rides["registered"]
registered_ma = registered.rolling(10).mean()
registered_ma_diff = registered - registered_ma
registered_ma_diff.dropna(inplace=True)
```

```python
casual = daily_rides["casual"]
casual_ma = casual.rolling(10).mean()
casual_ma_diff = casual - casual_ma
casual_ma_diff.dropna(inplace=True)
```

```python
# now test the stationarity of the two series
test_stationarity(registered_ma_diff, window=10, figsize=(20,10))
```

```python
test_stationarity(casual_ma_diff, window=10, figsize=(20,10))
```

```python
# subtract last value
registered = daily_rides["registered"]
registered_diff = registered - registered.shift()
registered_diff.dropna(inplace=True)
casual = daily_rides["casual"]
casual_diff = casual - casual.shift()
casual_diff.dropna(inplace=True)
```

```python
test_stationarity(registered_diff, figsize=(20,10))
```

```python
test_stationarity(casual_diff, figsize=(20,10))
```

## Time Series Decomposition in Trend, Seasonality,And Residual Components

```python
# Use the statsmodel.tsa.seasonal. seasonal_decompose() method
# to decompose the registered and casual rides
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition_reg = seasonal_decompose(daily_rides['registered'], model="additive")
decomposition_casual = seasonal_decompose(daily_rides['casual'], model="additive")

```

```python
registered_plot = decomposition_reg.plot()
registered_plot.set_size_inches(20,10)
casual_plot = decomposition_casual.plot()
casual_plot.set_size_inches(20,10)
```
